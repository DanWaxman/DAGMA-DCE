```python
def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)
```

```python
def make_box_plots(ax, v, c):
    isnan = ~np.isnan(v)
    res_filtered = [d[m] for d, m in zip(v, isnan)]
    bp = ax.boxplot(res_filtered, positions=np.array(range(len(res_filtered)))*2.0-0.4+0.4*(m-2), sym='', widths=0.3,
                    autorange=True, whis=((0, 100)))
    set_box_color(bp, c)
```
